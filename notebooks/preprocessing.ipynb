```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Читаем данные из _data.csv и выводим первые 5 строк dataframe

```python
df = pd.read_csv("/home/savely/Desktop/Data/apartment-price-analytics/data/_data.csv", index_col=0)
pd.set_option("display.max_columns", None)  # Показывать все столбцы
df.head(5)
```

#### Удаляем ненужные признаки из dataframe

```python
df.drop(["Тип", "Телефоны", "Описание", "Дополнительно", "Название ЖК", "Серия дома", "Ссылка на объявление"], axis=1, inplace=True)
df.head(3)
```

#### Оставляем строки, которые относятся только к городу Москва


```python
df = df[df["Адрес"].str.contains("Москва", case=False, na=False)]
df.head(3)
```

#### Переименовываем колонки на английский


```python
df = df.rename(columns={'ID  объявления': 'ad_id',
    'Количество комнат': 'num_rooms',
    'Метро': 'metro',
    'Адрес': 'address',
    'Площадь, м2': 'area_m2',
    'Дом': 'building',
    'Парковка': 'parking',
    'Цена': 'price',
    'Ремонт': 'renovation',
    'Площадь комнат, м2': 'room_area_m2',
    'Балкон': 'balcony',
    'Окна': 'windows',
    'Санузел': 'bathroom',
    'Можно с детьми/животными': 'pets_children_allowed',
    'Высота потолков, м': 'ceiling_height_m',
    'Лифт': 'elevator',
    'Мусоропровод': 'garbage_chute'})
df.head(3)
```

#### Удаляем пропущенные значения

```python
df = df.fillna(0)
df.isna().sum()
```

#### Корректируем значения в колонке "Цена"

```python
# Оставляем только первое число (если их несколько)
df["price"] = df["price"].astype(str).str.extract(r"^(\d+\.\d+|\d+)")[0]

# Преобразуем в float
df["price"] = df["price"].astype(float)
df.head(5)
```
