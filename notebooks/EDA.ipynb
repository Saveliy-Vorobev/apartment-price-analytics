### Проверяем пропущенные значения и смотрим их количество

### Количество городов/областей в dataframe

```python
df_city = list(df["Адрес"].apply(lambda x: x.split(",")[0]).unique())
df_city
```

```python
# Поскольку в нашем задании мы должны рассматривать только данные по городу Москва, то стоит уже на текущем этапе скорректировать dataframe и исключить данные о других городах
```

```python
df_moscow = df[df["Адрес"].str.contains("Москва", case=False, na=False)]
df_moscow.head(1)
```

### Количество уникальных значений у каждого признака

```python
# Смотрим на количество уникальных значений в каждом столбце, чтобы после можно было провести группировку при необходимости, а также на следующих релизах провести кодировку категориальных признаков.
# Помимо этого, мы также можем увидеть, что часть признаков являются неинформативными и их можно будет удалить в дальнейшем, например признак "Тип", поскольку он имеет только одно значение и не влияет на цену квартиры.
# Также признаки со слишком большим количеством уникальных значений тоже являются неинформативными, их можно будет удалить: Ссылка на объявление, Описание, Телефоны. 
df_moscow.nunique().reset_index()
```

### Проведение анализа зависимости признаков dataframe друг от друга
#### Корректируем тип данных у признаков

```python
# Для начала приведем колонку "Цена" к нужному типу данных, чтобы с ней можно было работать. # тоже касается остальных колонок, которые нам важны для определения различного рода зависимостей. 
# Приведем их к нужному типу данных

#Цена
df_moscow["Цена"] = df_moscow["Цена"].astype(str).str.extract(r"(\d[\d\s]*)").replace(" ", "", regex=True).astype(float)

# Количество комнат
df_moscow["Количество комнат"] = df_moscow["Количество комнат"].astype(str).str.extract(r"(\d+)")[0].astype(float)

# Площадь, м2
df_moscow["Площадь, м2"] = df_moscow["Площадь, м2"].astype(str).str.split("/").str[0].astype(float)

# Высота потолков
df_moscow["Высота потолков, м"] = df_moscow["Высота потолков, м"].astype(float)

```

#### График распределения цен на аренду в Москве

```python
# При построении гистограммы распределения цен, мы поставили ограничение на стоимость квартир, чтобы отбросить экстремально высокие цены. В целом, даже с процентиелм 99% мы видим, что квартир выше 450 тыс. рублей, 
# крайне мало на рынке объявлений, поэтому в дальнейшем, можно будет убрать из dataframe данные с ценами выше 450-500 тыс. рубле, чтобы выбрасы не влияли качественно на результат

q_high = df_moscow["Цена"].quantile(0.99)
df_filtered = df_moscow[df_moscow["Цена"] <= q_high]   # dataframe с ограничением цены. Цена без значительных выбросов

plt.figure(figsize=(10, 6))
sns.histplot(df_filtered["Цена"], bins=30, kde=True)
plt.xlabel("Цена аренды (руб.)")
plt.ylabel("Количество объявлений")
plt.title("Распределение цен на аренду")
plt.grid(True)
plt.savefig('/home/savely/Desktop/Data/apartment-price-analytics/tmp_data/delivery_price.png')
plt.show()
```

#### График зависимости цены от вида парковки

```python
# df_filtered["Парковка"].unique()
# Смотрим какие типы парков бывают. И поскольку мы знаем, что у признака "Парковка" весьма много пропущенных значений, то решаем заменить данные значения, на "бесплатная парковка"

df_filtered["Парковка"] = df_filtered["Парковка"].fillna("бесплатная парковка")

# и после этого строим график зависимости цены от вида парковки

plt.figure(figsize=(10, 6))
sns.barplot(data=df_filtered, x="Парковка", y="Цена", hue="Парковка", palette="cool")

# В целом можно считать, что данный признак является достаточно важным, и он демонстрирует, что от типа парковки цена достаточно сильно зависит. Также, видно, что все пропущенные значения, 
# которые представляют тип "бесплатная парковка" находятся примерно на одном уровне с типом парковки: "наземная", что говорит о том, что пропущенные значения скорее всего и являлись этим типом. 
# Поэтому в дальнейшем замену пропущенных значений можно проводить именно по этому типу
```

#### График зависимости цены аренды от типа дома, количества этажей и этажа квартиры

```python
# Создадим дополнительные признаки "Тип дома", "Количество этажей" и "Этаж квартиры" из переменной "Дом"
def get_house_type(x):
    if isinstance(x, str):
        try:
            return x.split('/')[1].split(', ')[1]
        except:
            return np.nan
    return x

def get_total_floors(x):
    if isinstance(x, str):
        return int(x.split('/')[1].split(',')[0])
    return x

def get_floor(x):
    if isinstance(x, str):
        return int(x.split('/')[0])
    return x

df_filtered['Тип дома'] = df_filtered['Дом'].map(get_house_type)
df_filtered['Количество этажей'] = df_filtered['Дом'].map(get_total_floors)
df_filtered['Этаж квартиры'] = df_filtered['Дом'].map(get_floor)

# А теперь посмотрим зависимость цены от новых признаков
plt.figure(figsize=(12, 6))
sns.barplot(data=df_filtered, x="Тип дома", y="Цена", palette="cool")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.relplot(x=df_filtered["Этаж квартиры"], y=df_filtered["Цена"], hue=df_filtered["Количество этажей"], col=df_filtered["Тип дома"], col_wrap=3)
plt.show()
```

#### График зависимости цены аренды от ближайшего метро

```python
df_filtered["Метро"] = df_filtered["Метро"].str.extract(r'м\. ([^(]+)')
df_filtered["Метро"]
df_filtered = df_filtered.dropna(subset=["Метро"])

plt.figure(figsize=(14, 6))
sns.barplot(data=df_filtered, x="Метро", y="Цена", estimator=np.mean, hue="Метро", palette="viridis", legend=False)

# Настройки графика
plt.xticks(rotation=90)
plt.xlabel("Станция метро")
plt.ylabel("Средняя цена квартиры")
plt.title("Зависимость цены квартиры от станции метро")

plt.show()
```
